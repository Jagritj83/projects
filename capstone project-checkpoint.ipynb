```python
df1['Murder'] = df1['Murder'].fillna(df1.groupby('STATE/UT')['Murder'].transform('median'))
df1['Rape'] = df1['Rape'].fillna(df1.groupby('STATE/UT')['Rape'].transform('mean'))
df1['Kidnapping and Abduction'] = df1['Kidnapping and Abduction'].fillna(df1.groupby('STATE/UT')['Kidnapping and Abduction'].transform('median'))
df1['Foeticide'] = df1['Foeticide'].fillna(df1.groupby('STATE/UT')['Foeticide'].transform('median'))
df1['Abetment of suicide'] = df1['Abetment of suicide'].fillna(df1.groupby('STATE/UT')['Abetment of suicide'].transform('median'))
df1['Exposure and abandonment'] = df1['Exposure and abandonment'].fillna(df1.groupby('STATE/UT')['Exposure and abandonment'].transform('median'))
df1['Procuration of minor girls'] = df1['Procuration of minor girls'].fillna(df1.groupby('STATE/UT')['Procuration of minor girls'].transform('median'))
df1['Buying of girls for prostitution'] = df1['Buying of girls for prostitution'].fillna(df1.groupby('STATE/UT')['Buying of girls for prostitution'].transform('median'))
df1['Selling of girls for prostitution'] = df1['Selling of girls for prostitution'].fillna(df1.groupby('STATE/UT')['Selling of girls for prostitution'].transform('median'))
df1['Prohibition of child marriage act'] = df1['Prohibition of child marriage act'].fillna(df1.groupby('STATE/UT')['Prohibition of child marriage act'].transform('median'))
df1['Other Crimes'] = df1['Other Crimes'].fillna(df1.groupby('STATE/UT')['Other Crimes'].transform('median'))
```
